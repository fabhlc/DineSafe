Save raw geocoded file to txt file for retrieval later.

```python
with open("geocoded.txt", 'w') as f:
    for i in results_list:
        f.write(str(i))
```

Parse out from raw geocoded into a clean dataframe.

```python
geo_df = pd.DataFrame(columns = ['address', 'lat','lon'])
for i in results_list:
    address = i[0]['address_components'][0]['short_name'] + ' ' + i[0]['address_components'][1]['short_name'].upper()
    lat = i[0]['geometry']['location']['lat']
    lon = i[0]['geometry']['location']['lng']
    
    mini_df = pd.DataFrame(data=[[address, lat,lon]], columns = ['address', 'lat','lon'])
    geo_df = geo_df.append(mini_df)

geo_df['original_address'] = list(to_geocode)

geo_df
```

Save to CSV for later retrieval.

```python
geo_df.to_csv('geocoded_df.csv', index = None)
```
