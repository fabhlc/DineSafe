### DATA PRE-PROCESSING PART 1
First notebook in our DineSafe Gentrification project which takes the raw data and combines it. End result is a dataframe that shows only one instance of Establishment_Id per year from 2012 to 2018-ytd.

#### Six Dinesafe files from the following dates (downloaded via Web Archive version of Open Data Toronto) are combined:
- July 2013
- February 2015
- January 2016
- September 2016
- March 2017
- November 2018

```python
from collections import Counter
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import time

# Set display options
pd.set_option("display.max_rows", 500)

data_dir = r'C:\Users\Fabienne\Documents\Datasets\City of Toronto - Dinesafe\\'

# xmlp = ET.XMLParser(encoding="ANSI")

# Load most recent Dinesafe Data (Dec 12th 2018):
with open(data_dir + 'dinesafe_12Dec2018.xml','r') as f:
    ef = ET.fromstring(f.read())
    rows = ef.findall("ROW")

ds_dict = dict()

for num, i in enumerate(rows):
    ds_dict[num] = {}
    children = i.getchildren()
    for j in children:
        ds_dict[num][j.tag] = j.text

ds_df = pd.DataFrame.from_dict(ds_dict, orient="index")
del ds_dict, rows
ds_df
```

```python
# Load previous Dinesafe data

# March 2017
with open(data_dir + 'Old Dinesafe/dinesafe_Mar2017.xml','r') as f:
    mar2017 = ET.fromstring(f.read())
    mar2017_rows = mar2017.findall("ROW")
mar2017_dict = dict()
for num, i in enumerate(mar2017_rows):
    mar2017_dict[num] = {}
    children = i.getchildren()
    for j in children:
        mar2017_dict[num][j.tag] = j.text
mar2017_ds = pd.DataFrame.from_dict(mar2017_dict, orient="index")
del mar2017_dict, mar2017_rows

# January 2016
with open(data_dir + 'Old Dinesafe/dinesafe_Jan2016.xml','r') as f:
    jan2016 = ET.fromstring(f.read())
    jan2016_rows = jan2016.findall("ROW")
jan2016_dict = dict()
for num, i in enumerate(jan2016_rows):
    jan2016_dict[num] = {}
    children = i.getchildren()
    for j in children:
        jan2016_dict[num][j.tag] = j.text
jan2016_ds = pd.DataFrame.from_dict(jan2016_dict, orient="index")
del jan2016_dict, jan2016_rows

# September 2016
with open(data_dir + 'Old Dinesafe/dinesafe_Sep2016.xml','r') as f:
    sep2016 = ET.fromstring(f.read())
    sep2016_rows = sep2016.findall("ROW")
sep2016_dict = dict()
for num, i in enumerate(sep2016_rows):
    sep2016_dict[num] = {}
    children = i.getchildren()
    for j in children:
        sep2016_dict[num][j.tag] = j.text
sep2016_ds = pd.DataFrame.from_dict(sep2016_dict, orient="index")
del sep2016_dict, sep2016_rows

# February 2015
with open(data_dir + 'Old Dinesafe/dinesafe_Feb2015.xml','r') as f:
    feb2015 = ET.fromstring(f.read())
    feb2015_rows = feb2015.findall("ROW")
feb2015_dict = dict()
for num, i in enumerate(feb2015_rows):
    feb2015_dict[num] = {}
    children = i.getchildren()
    for j in children:
        feb2015_dict[num][j.tag] = j.text
feb2015_ds = pd.DataFrame.from_dict(feb2015_dict, orient="index")
del feb2015_dict, feb2015_rows

# July 2013
with open(data_dir + 'Old Dinesafe/dinesafe_Jul2013.xml','r') as f:
    jul2013 = ET.fromstring(f.read())
    jul2013_rows = jul2013.findall("ROW")
jul2013_dict = dict()
for num, i in enumerate(jul2013_rows):
    jul2013_dict[num] = {}
    children = i.getchildren()
    for j in children:
        jul2013_dict[num][j.tag] = j.text 
jul2013_ds = pd.DataFrame.from_dict(jul2013_dict, orient="index")
del jul2013_dict, jul2013_rows
```

```python
print(sorted(Counter([i[:7] for i in jul2013_ds['INSPECTION_DATE']])))
print(sorted(Counter([i[:7] for i in feb2015_ds['INSPECTION_DATE']])))
print(sorted(Counter([i[:7] for i in jan2016_ds['INSPECTION_DATE']])))
print(sorted(Counter([i[:7] for i in sep2016_ds['INSPECTION_DATE']])))
print(sorted(Counter([i[:7] for i in mar2017_ds['INSPECTION_DATE']])))
print(sorted(Counter([i[:7] for i in ds_df['INSPECTION_DATE']])))
```

Inspections are not consistent across datasets - suspect there may be a live-feed component to it. Regardless, we'll filter unique INSPECTION_IDs. 


Then to find unique establishments we'll filter out ESTABLISHMENT_ID.

```python
ds = ds_df.append(mar2017_ds.append(sep2016_ds.append(jan2016_ds.append(feb2015_ds.append(jul2013_ds)))))
len_1 = len(ds)
print(ds.info())
# Counter(ds['INSPECTION_ID']).most_common()
ds = ds.drop_duplicates("INSPECTION_ID")
print("{:,} duplicates removed. Length changed from {:,} to {:,}".format(len_1-len(ds), len_1, len(ds)))
```

```python
print("Establishment IDs that don't have lat/lon:", len(set(ds[ds['LATITUDE'].isnull()]['ESTABLISHMENT_ID'])))
print("Total length:", len(set(ds['ESTABLISHMENT_ID'])))
```

```python
# Drop unnecessary columns
ds.drop(['ACTION', 'AMOUNT_FINED', 'COURT_OUTCOME', 'ROW_ID','SEVERITY', 'ESTABLISHMENT_STATUS','INFRACTION_DETAILS'], axis = 1, inplace = True)

# Strip whitespace from all columns
ds = ds.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

ds.sample(15)
```

### CREATE REFERENCE DATASETS FOR LAT AND LONG
#### Out of One Address and most recent 2018 dinesafe file 

```python
# 1.) One Address
one_add_dir = r'C:\Users\Fabienne\Documents\Datasets\City of Toronto - One Address repository\Dec 2018 One Address (as of Sep 2017)\OneAddress_2017.csv'
with open(one_add_dir, 'r') as f:
    one_add = pd.read_csv(f)
pd.set_option("display.max_columns", 50)

one_add['street'] = one_add['ADDRESS'] + ' ' + one_add['LFNAME'].str.upper()

one_add = one_add[['street', 'LONGITUDE', 'LATITUDE']]


# 2.) Most recent Dnesafe file (ds_df)
adddict_2018 = ds_df.drop_duplicates('INSPECTION_ID')
adddict_2018 = adddict_2018[['ESTABLISHMENT_ADDRESS', 'LATITUDE','LONGITUDE']]


# Combine the two reference datasets.
one_add.columns = ['ESTABLISHMENT_ADDRESS','LONGITUDE', 'LATITUDE']
address_dict = one_add.append(adddict_2018)
del one_add, adddict_2018

# Remove duplicates
address_dict = address_dict.drop_duplicates('ESTABLISHMENT_ADDRESS')

address_dict.sample(30)
```

```python
# Strip whitespace
ds['ESTABLISHMENT_ADDRESS'] = ds['ESTABLISHMENT_ADDRESS'].str.strip()
address_dict['ESTABLISHMENT_ADDRESS'] = address_dict['ESTABLISHMENT_ADDRESS'].str.strip()

# Shorten column name
ds = ds.rename(columns = {'MINIMUM_INSPECTIONS_PERYEAR':'risk_level'})

# Join to address dictionary
ds = ds.join(address_dict.set_index("ESTABLISHMENT_ADDRESS"), on="ESTABLISHMENT_ADDRESS", how = "left" ,rsuffix="_join")

no_geo = ds[ds['LONGITUDE_join'].isnull()]
print("{:.2f} % of establishments have lat and lon data".format((sum(ds['LONGITUDE_join'].notnull())/len(ds))*100))
ds.sample(30)
```

```python
ds['Year'] = [int(i[:4]) for i in ds['INSPECTION_DATE']]

# Sort by Establishment ID
ds = ds.sort_values('ESTABLISHMENT_ID')

# Keep only one instance of Establishment_Id per year
ds = ds.drop_duplicates(['ESTABLISHMENT_ID','Year'])

# Reset Index
ds.reset_index(drop=True, inplace=True)

# Replace NaN in Lat/Lon with joined Lat/Lon
ds['LATITUDE'].fillna(ds['LATITUDE_join'], inplace = True)
ds['LONGITUDE'].fillna(ds['LONGITUDE_join'], inplace = True)
ds.drop(['LATITUDE_join','LONGITUDE_join'], axis = 1, inplace = True)
```

#### Export a list of remaining addresses which need to be geocoded (separate notebook).

```python
to_geocode = ds[ds['LATITUDE'].isnull()]['ESTABLISHMENT_ADDRESS'].unique()
print(len(to_geocode), "\n", to_geocode)
```

## Geocode Remaining Addresses
Import leftover addresses (183 unique) which were geocoded using Google Geocoding API (see separate Notebook for process).

```python
with open("Data/geocoded_df.csv", 'r') as file:
    geo_df = pd.read_csv(file, index_col = None)
```

```python
set(to_geocode) - set(geo_df['original_address'])
```

```python
# Join 
ds = ds.join(geo_df.set_index('original_address'), on = "ESTABLISHMENT_ADDRESS", how = "left", rsuffix = '_geocoded')

# Replace NaN in lat/lon with geocoded lat lon and drop geocoded column after
ds['LATITUDE'].fillna(ds['lat'], inplace = True)
ds['LONGITUDE'].fillna(ds['lon'], inplace = True)
ds.drop(['lat', 'lon','address'], axis = 1, inplace = True)

# Remove incomplete 2011 data
ds = ds[ds['Year'] != 2011]
ds = ds.sort_values(by = ['ESTABLISHMENT_ID', 'Year'])
ds.drop('INSPECTION_DATE', axis = 1, inplace = True)

ds.reset_index(drop=True, inplace = True)

# Reorder columns to show Year first
cols = ds.columns.tolist()
cols = cols[-1:] + cols[:-1]
ds = ds[cols]

ds
```
